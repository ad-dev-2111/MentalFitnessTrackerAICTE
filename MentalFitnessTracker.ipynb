## split the data into input and output variable

```python
# define the dependent and independent variable
X = df.drop('Mental_fitness',axis=1)
X = df.drop('Country',axis=1)
y = df['Mental_fitness']
```

## Scaling the data

```python
# Scale the data to bring all the attributes to a comparable level
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Spliting the data into train test

```python
# split the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.20,random_state=2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## Importing and working of LinerRegression model on our data

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
# create a dataframe for actual and predicted values
temp = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
temp.head()
```

## Showing the difffrence between Actual and predicted using Graphs

```python
# show the actual and predicted value in graph
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.scatterplot(x='Actual',y='Predicted',data=temp)
sns.lineplot(x='Actual',y='Actual',data=temp,color='r')
plt.subplot(1,2,2)
sns.distplot(temp['Actual'],color='r')
sns.distplot(temp['Predicted'],color='b')
```

## Showing the values related to model matrices

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)   # calculate the mean squared error
rmse = np.sqrt(mse)   # calculate the root mean squared error

print('Mean Squared Error:',mse)
print('Root Mean Squared Error:',rmse)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)   # calculate the r2 score
print('R2 Score:',r2)
```

```python
# accuracy of the model
print('Accuracy of the model:',lr.score(X_test,y_test))
```
